 extraction de la durée min/max/moyenne par trajet

    Avant d'éffectuer nos calcul, il est important de vérifier le type des données notemment de la colonne date

```python
frame_ticket.info()
```

    Nous remarquons que les dates de départs et d'arrivées sont de types object. Nous allons donc les convertir en datetime 
    et ensuite nous allons ajouter une nouvelle colonne devant contenir la difference de temps, que nous appelerons         'traject_ts'

```python
frame_ticket['arrival_ts'] = pd.to_datetime(frame_ticket['arrival_ts'])
frame_ticket['departure_ts'] = pd.to_datetime(frame_ticket['departure_ts'])
frame_ticket['traject_ts'] = frame_ticket['arrival_ts'] - frame_ticket['departure_ts']
```

     Nous pouvons ensuite convertir cette difference de dates en secondes afin d'avoir une valeur numérique
     Pour cela nous allons créer une nouvelle colonne appelée 'diff_seconds' 

```python
frame_ticket['diff_seconds'] = frame_ticket['traject_ts'].dt.total_seconds()
```

    Enfin, nous pouvons extraire de la durée min/max/moyenne par trajet

```python
df_groupBy_trajet['diff_seconds'].agg([np.min, np.mean, np.max])
```

    Ajout de la colonne 'distance' qui contiendra la distance entre 2 stations:
        Nous allons évaluer cette distance par vol d'oiseau en utilisant 
        les longitudes et latitudes des stations d'origine et destination

```python
       Avant de calculer la distance nous allons effectuer 2 jointures successives afin d'y ajouter les 
       coordonnées géographiques des stations d'origine et de destination au tickets.
```

```python
ticket_stationOrigine=pd.merge(frame_ticket, frame_stations, left_on="o_station", right_on="id", how="inner")
ticket_stationOrigineDestination=pd.merge(ticket_stationOrigine, frame_stations, left_on="d_station", right_on="id", how="inner")
```

```python
ticket_stationOrigineDestination.head(5)
```

```python
import math as math
from math import radians, cos, sin, asin, sqrt
def distance(lat1, lng1, lat2, lng2):
    #return distance a km distance
    radius = 6371  

    dLat = (lat2-lat1) * math.pi / 180
    dLng = (lng2-lng1) * math.pi / 180

    lat1 = lat1 * math.pi / 180
    lat2 = lat2 * math.pi / 180

    val = sin(dLat/2) * sin(dLat/2) + sin(dLng/2) * sin(dLng/2) * cos(lat1) * cos(lat2)    
    ang = 2 * math.atan2(sqrt(val), sqrt(1-val))
    return radius * ang
```

```python
ticket_stationOrigineDestination['distance'] = ticket_stationOrigineDestination.apply(lambda x: distance(x.latitude_x,x.longitude_x,x.latitude_y,x.longitude_y), axis=1)
```

```python
ticket_stationOrigineDestination.head(5)
```

      Regroupement des distance en fonction des intervalles: 0-200km, 201-800km, 800-2000km, 2000+km

```python
#ticket_stationOrigineDestination["distance"]=pd.cut(ticket_stationOrigineDestination["distance"], bins=[0,200,800,2000,ticket_stationOrigineDestination["distance"].max()],include_lowest = True)
ticket_stationOrigineDestination["distance"]=pd.cut(ticket_stationOrigineDestination["distance"], bins=4,include_lowest = True)
```

    Pour savoir le moyen de transport(train, bus ou covoi) nous devons faire une jointure avec le dataframe 'frame_providers'

```python
ticket_stationOrigineDestination_provider = pd.merge(ticket_stationOrigineDestination, frame_providers, left_on="company", right_on="company_id", how="inner")
ticket_stationOrigineDestination_provider.head(2)
```

   Différence de prix moyen selon le train, le bus et le covoit selon la distance du trajet

```python
ticket_stationOrigineDestination_provider.pivot_table(values='price_in_cents', index='distance', columns='transport_type',aggfunc='mean')
```

Différence de durée moyenne selon que l'on emprunte le train, le bus et le covoit et selon la distance du trajet

```python
ticket_stationOrigineDestination_provider.pivot_table(values='diff_seconds', index='distance', columns='transport_type',aggfunc='mean')
```
